```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("dark")
%matplotlib inline
x = np.arange(100)
```

```python
data = pd.read_csv("data/drug-use-by-age.csv")
```

```python
ax, fig = plt.subplots()
plt.plot(data["age"][:8], data["marijuana-use"][:8], color="red")
plt.plot(data["age"][:8], data["cocaine-use"][:8], color="blue")
plt.title("Drug use by Age")
plt.xlabel("Age")
plt.ylabel("Percentage of drug users")
plt.legend()
```
