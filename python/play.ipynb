```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("dark")
%matplotlib inline
x = np.arange(100)
```

```python
data = pd.read_csv("data/drug-use-by-age.csv")
mj_use = data["marijuana-use"][:8]
age = data["age"][:8]
coke_use = data["cocaine-use"][:8]
```

```python
plt.plot(age, mj_use)
plt.plot(age, coke_use)
plt.legend()
plt.xlabel("Age")
plt.ylabel("Percentage of drug users")
plt.title("Drug use by Age")
```

```python
def sine():
    x = np.linspace(0, 10, 100)
    for i in range(10):
        plt.plot(x, np.sin(x + i * .25))
    plt.plot(x, x * 0, linestyle="dashed")
```

```python
sine()
```
